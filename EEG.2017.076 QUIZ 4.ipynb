```python
import matplotlib.pyplot as plt
from math import pi, sqrt

Voltage=[-157.979 * i/50 for i in range(50, -1, -1)]
Frequency = [1/(2*pi * sqrt( 5e-9 * 1.3702e-3 * 3.6972e-9 * sqrt(1- V/0.6))) for V in Voltage]

fig, graph = plt.subplots()
graph.plot(Voltage, Frequency)

graph.set(xlabel='Voltage-mV', ylabel='Frequency-Hertz')
graph.grid()
```
